```python
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.106 Safari/537.36'}
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.ameximoveis.com.br/categoria')
time.sleep(5)
element = driver.find_element(By.XPATH,'//footer')
previous_height = driver.execute_script('return document.body.scrollHeight')
#print(previous_height)

while driver.find_element(By.XPATH,'//footer'):
    driver.execute_script("arguments[0].scrollIntoView();",element)
    #divs = driver.find_element(By.XPATH,'//footer').text
    time.sleep(5)
    new_height = driver.execute_script('return document.body.scrollHeight')
    #print(new_height)
    if new_height <= previous_height:
        resposta = driver.page_source
        site = BeautifulSoup(resposta, 'html.parser')
        print(site.prettify())
        close.browser()
        #print('OK - Break')
        break
    else:
        #print('OK - Continue')
        previous_height = new_height
        continue
```

```python
price_list = site.find_all('div','price')
#print(price_list)

for price in price_list:
    price = price.find('h5').text.replace('Venda R$ ','').replace('.','')
    print(price)
```

```python
type_list = site.find_all('div','badges')
#print(price_list)

for type_ in type_list:
    type_ = type_.find('span').text
    print(type_)
```

```python
list_spec = []

spec_list_num = site.find_all('div','spec-item')
#print(price_list)

for spec in spec_list_num:
    spec_num = spec.find('span').text.replace(' quartos','').replace(' suítes','').replace(' m²','').replace(' banheiros','').replace(' garagens','').replace('.','').replace(' suíte','').replace(' banheiro','').replace(' garagem','').replace(' quarto','')
    list_spec.append(spec_num)

list_spec
```

```python
x = list_spec
mat = []
while x != []:
  mat.append(x[:5])
  x = x[5:]
df = pd.DataFrame(mat, columns=['m²','quarto','suíte','banheiro','garagem'])
df
```

```python
# determining the name of the file
file_name = 'specs.xlsx'
  
# saving the excel
df.to_excel(file_name)
```
