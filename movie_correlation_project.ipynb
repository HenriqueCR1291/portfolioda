```python
df1 = df.dropna()
```

```python
#data type
df1.dtypes
```

```python
#change data type
df1['budget'] = df1['budget'].astype('int64')

df1['gross'] = df1['gross'].astype('int64')
```

```python
#correct the year

#df1.dtypes
#df1['yearcorrect'] = re.search('\d\S([0-9])\w', str(df1['yearcorrect']))

#df1['yearcorrect'] = df1['released'].apply(lambda x: re.findall('[0-9]{4}', x))
df1['yearcorrect'] = df1['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
#df1['yearcorrect'] = df1['yearcorrect'].astype('int64')
df1.head(3)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df1.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#correlation matrix

correlation_matrix = df1.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Metric for movies')

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
df_num = df1

for col_name in df_num.columns:
    if (df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes
        
df_num
```

```python
correlation_matrix = df_num.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Metric for movies')

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
correlation_mat = df_num.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs = sorted_pairs[((sorted_pairs) >= 0.5) & ((sorted_pairs) < 1.0)]
sorted_pairs
```
